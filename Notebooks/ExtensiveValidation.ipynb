```r
library(tidyverse)
```

```r
path = "/home/and/RIDIR/Code/Validation/"
dataset = "Source2000/"
spark_extensive = read_tsv(paste0(path, dataset, "geospark_extensive.tsv"), col_names = F) %>%
    rename(tid = X1, spark = X2) %>%
    arrange(tid)

spark_extensive %>% show
```

```r
pandas_extensive = read_tsv(paste0(path, dataset, "geopandas_extensive.tsv"), col_names = F) %>%
    rename(tid = X1, pandas = X2) %>%
    arrange(tid)

pandas_extensive %>% show
```

```r
joined = pandas_extensive %>% inner_join(spark_extensive, by="tid")
joined %>% show
```

```r
ggplot(data = joined, aes(x = pandas, y = spark)) + geom_point()
```

```r
library(MLmetrics)

print(paste("R2:  ", R2_Score(joined$spark, joined$pandas)))
print(paste("RMSE:", RMSE(joined$spark, joined$pandas)))
print(paste("MAE: ", MAE(joined$spark, joined$pandas)))
print(paste("MSE: ", MSE(joined$spark, joined$pandas)))
```
